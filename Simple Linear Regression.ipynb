## splitting dataset into X ( Feature - Independent Variable) and y (target - Dependent Variable)


y = b1 * X + b0

y = Target = Dependent variable 

b1 = slope / gradient / coefficient   --> -1 < b1 < 1,  b1 != 0

X = Featue = Independent Variable 

b0 = Intercept = Otonom Parameter = start parameter --> - ∞ < b0 < +∞


```python
X = df["total_spend"]
y = df["sales"]
```

```python
np.polyfit(X, y, deg = 1)
```

### Determination of Coefficients (Slope and Intercept)

```python
slope, intercept = np.polyfit(X, y, deg = 1)
```

```python
print(f"slope        : {slope}")
print(f"intercept    : {intercept}")
```

```python
b1 ,b0 = np.polyfit(X, y, deg = 1)
```

```python
print(f"b1        : {slope}")
print(f"b0        : {intercept}")
```

### y_pred Formula

y_pred = b1 * X + b0

```python
y_pred = b1 * X + b0
y_pred
```

```python
df.head()
```

```python
values = { "actual": y, "predicted": y_pred, "residual": y-y_pred, "LSE": (y-y_pred)**2}
df_2 = pd.DataFrame(values)
df_2.head(20)
```

LSE is calculated since negative values cancel positives and correct calculation cant be derived. Thats why, squared values are preferred.

best fitted line is drawn where the residuals' sum is close to 0 as much as possible.

```python
df_2.residual.sum()
```

```python
df_2.residual.sum().round()
```
